```python
import warnings
warnings.filterwarnings('ignore')
from darkflow.net.build import TFNet
```

```python
options={'model':'cfg/rc.cfg','load':4000,'threshold':0,'gpu':0}
tfnet=TFNet(options)
```

```python
import cv2
import numpy as np
import time
import pafy
from tensorflow.keras.preprocessing.image import load_img
```

```python
colors=[tuple(255*np.random.rand(3)) for _ in range(10)]
```

```python
import urllib
from urllib import request as ur
from datetime import datetime
url = "https://www.youtube.com/watch?v=-mlr860BSvk"
videoPafy = pafy.new(url)
best = videoPafy.getbest()
capture=cv2.VideoCapture(0)
import os
import matplotlib.pyplot as plt
```

```python
count=0
path=os.listdir('D:/data/rc test/')
for x in path:
    img=load_img('D:/data/rc test/'+x)
    img=np.array(img)
    result=tfnet.return_predict(img)
    for color,result in zip(colors,result):
        
        if result['label']=='registration number':
            try:
                x1,y1=result['topleft']['x'],result['topleft']['y']
                x2,y2=result['bottomright']['x']+200,result['bottomright']['y']+20
                img=img[y1:y2,x1:x2]
                plt.imshow(img)
                cv2.imwrite('D:/data/rc prediction output/registration number/'+x,img)
                count+=1
            except:
                print(x,'out of bounds.')
     
```
